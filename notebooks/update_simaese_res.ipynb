```python
import pandas as pd

SUB = '../submissions/simaese_mob_299_sim_best_train_loss_2019-02-24_11_42.csv'
old_df = pd.read_csv(SUB)
print(len(old_df))
old_df.head(10)
```

```python
new_id = []
# new whale should in top2
for old_id in old_df.Id.values:
    tmp_l = old_id.split(' ')
    if 'new_whale' not in tmp_l:
        tmp_l.insert(1, 'new_whale')
    elif 'new_whale' in tmp_l[-3:]:
        tmp_idx = tmp_l.index('new_whale')
        tmp_l[1], tmp_l[tmp_idx] = tmp_l[tmp_idx], tmp_l[1]
    else:
        pass
    new_str = ' '.join(tmp_l[:5])
    new_id.append(new_str)
new_df = old_df.copy()
new_df.Id = new_id
new_df.head(10)
```

```python
new_df.to_csv(SUB.replace('submissions/','submissions/update_top2_'), index=False) # still bad
print('save done')
```
